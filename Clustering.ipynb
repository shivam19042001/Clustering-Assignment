1.What is unsupervised learning in the context of machine learning

```python
"""
Unsupervised learning lets a model explore raw data with no provided answers or labels. Instead of
being told what to look for, it discovers patterns, groupings, or structure all on its own.

Common uses include:

-Clustering: Grouping similar items together (think: organizing customers by what they buy).

-Dimensionality reduction: Boiling down lots of data into fewer key features (making it simpler to visualize or work with).

-Anomaly detection: Spotting unusual entries that stick out from the rest (like fraud or errors).
"""
```

2.How does K-Means clustering algorithm work

```python
"""
K‑Means starts by picking k random points (the “centroids”) as the seeds for clusters. It then repeatedly follows two steps:

1.Assign each data point to the nearest centroid.

2.Update each centroid to be the average position of all points assigned to it.

These two steps—assign and update—keep repeating until the centroids stop moving much (they settle into place)
or you hit a set limit on iterations
"""
```

3.Explain the concept of a dendrogram in hierarchical clustering

```python
"""
A dendrogram is basically a tree-shaped chart that shows how groups merge in hierarchical clustering:

-At the bottom, every data point starts in its own little group.

-As you move upward, the closest pairs are merged step by step, forming larger clusters.

-The height of each join (vertical line) shows how different those clusters were—short low merges mean very
 similar groups, while tall high merges mean more dissimilar ones.
"""
```

4.What is the main difference between K-Means and Hierarchical Clustering

```python
"""
K‑Means vs. Hierarchical Clustering — simplified comparison

-->K‑Means

-You choose the number of clusters (k) up front.

-It partitions the data into flat, non-overlapping groups—no structure or nesting.

-It’s fast and works well on large datasets, but you need to know or guess how many groups to look for.

-->Hierarchical Clustering

-You don’t need to decide the number of clusters in advance—you build a full “family tree” of clusters.

-This tree (dendrogram) lets you see how clusters relate and choose the right cut based on data insight.

-It’s easier to interpret but can be slower and more resource-intensive, especially on large datasets.


"""
```

5.What are the advantages of DBSCAN over K-Means

```python
"""
DBSCAN (Density-Based Clustering)

-No need to pick cluster count ahead of time: It figures out how many groups naturally exist.

-Handles any cluster shape: Can form odd, elongated, or complex shapes—unlike K‑Means, which tends to make round clusters.

-Resists noise and outliers: It ignores scattered points as “noise,” instead of forcing them into a cluster.
"""
```

6.When would you use Silhouette Score in clustering

```python
"""
Silhouette Score is a way to check how well a point belongs in its cluster:

-It compares how close a point is to others in the same cluster (cohesion) with how far it is
 from points in the nearest other cluster (separation).

-Scores range from –1 to +1:

  +1: Perfect fit—close to its cluster and far from others.

  0: On the border between clusters.

  –1: Likely placed in the wrong cluster.
"""
```

7.What are the limitations of Hierarchical Clustering

```python
"""
Hierarchical clustering:

-Becomes slow and memory-heavy on large datasets because it calculates distances between all pairs of points.

-Is easily influenced by noise or outliers, which can skew the way clusters are formed.

-Is rigid: once it merges or splits clusters, it can’t revise those decisions later, making it less
 flexible if initial groupings were off.
"""
```

8.Why is feature scaling important in clustering algorithms like K-Means

```python
"""
K‑Means relies on calculating distances (usually Euclidean), so if one feature has values that range much larger than others,
it'll disproportionately influence the clustering results. That’s why scaling—bringing all features to a comparable range—is
essential. It ensures each feature has a fair impact, leading to more balanced clusters and faster, more stable convergence.
"""
```

9.How does DBSCAN identify noise points

```python
"""
DBSCAN labels as noise any point that doesn’t have a minimum number of neighbors (the min_samples threshold)
within a certain distance (eps). If a point doesn’t meet that density requirement, it’s kept out of all clusters.
"""
```

10.Define inertia in the context of K-Means

```python
"""
Inertia represents the overall tightness of clusters in K‑Means—it’s calculated by squaring the distance from
each point to its cluster center and then summing all those values. A lower inertia score means clusters are more compact and cohesive.
"""
```

11.What is the elbow method in K-Means clustering

```python
"""
The elbow method works like this:

-You run K‑Means with different values of k (e.g., 1 through 10).

-For each k, you calculate the total of squared distances between points and their centroids (inertia/WCSS).

-You then plot these values against k.

What you'll typically see is:

-A sharp drop in inertia as k increases, until a certain point.

-After that, the decrease slows down significantly, forming a visible bend or “elbow.”

-That elbow point—where the curve flattens—is a good guess for the optimal number of clusters.

In other words, it's where adding more clusters stops giving meaningful improvements.
"""
```

12.Describe the concept of "density" in DBSCAN

```python
"""
In DBSCAN:

-Density refers to how many surrounding points there are within a radius eps.

-If a point has at least min_samples within that radius, it’s considered a core point.

-Core points act as the foundation of clusters by linking together nearby dense regions.
"""
```

13.Can hierarchical clustering be used on categorical data

```python
"""
When applying hierarchical clustering to categorical data, you can’t rely on ordinary distances like Euclidean.
Instead, you need metrics tailored for categories—such as Hamming distance for pure categorical data or Gower
distance when mixing numeric and categorical variables. Without using these, standard linkage approaches can
produce misleading groupings.
"""
```

14.What does a negative Silhouette Score indicate

```python
"""
A negative Silhouette Score means a data point is likely in the wrong cluster. In other words, it’s closer
on average to points in a different cluster than to those in its own, suggesting its current grouping may
be inappropriate.
"""
```

15.Explain the term "linkage criteria" in hierarchical clustering

```python
"""
Linkage criteria in hierarchical clustering define how you measure the distance between two clusters:

1.Single linkage (nearest neighbor):
Uses the smallest distance between any two points—one from each cluster. This can result in long, “chain-like” clusters.

2.Complete linkage (farthest neighbor):
Relies on the largest distance between points in the clusters, producing tighter and more compact groupings.

3.Average linkage (group average):
Calculates the mean of all pairwise distances between points in the two clusters, creating a middle ground between single and complete linkage.
"""
```

16.Why might K-Means clustering perform poorly on data with varying cluster sizes or densities

```python
"""
K‑Means assumes that all clusters are roughly the same size, density, and shape—basically, ball-like.
When clusters vary in shape or density:

-It can mistakenly group far-off points together.

-Dense clusters may be split awkwardly.

-Sparse or oddly shaped clusters get misrepresented.

In other words, K‑Means works best when clusters are uniform and circular. If your data includes varied
densities or irregular shapes, it’s likely to miscluster points or fracture natural groups.
"""
```

17.What are the core parameters in DBSCAN, and how do they influence clustering

```python
"""
DBSCAN defines a neighborhood around each point using eps—any point within that radius is considered close.
If a point has at least min_samples neighbors inside this neighborhood, it's marked as a core point, which
becomes the center of a cluster. Increasing eps widens the neighborhood, creating bigger clusters. Raising
min_samples means you need a denser core for clusters to form, making it harder to create clusters unless points are tightly packed.
"""
```

18.How does K-Means++ improve upon standard K-Means initialization

```python
"""
K-Means++ improves clustering by choosing smarter starting points:

1.It selects the first centroid randomly.

2.Each subsequent centroid is picked with higher probability for data points that are farthest from
 already chosen centroids—using a distance-squared weighting.

3.This ensures the initial centroids are spread out, avoiding clumps and overlapping starts.

Why it helps:

-Reduces the chance of poor local optima—centroids start in better places.

-Decreases the number of iterations needed—converges faster.

-Results in more reliable and often more accurate clustering than random initialization.
"""
```

19.What is agglomerative clustering

```python
"""
Agglomerative clustering, a “bottom-up” hierarchical method, begins with each data point treated
 as its own cluster. Then, it repeatedly merges the two closest clusters (based on your chosen
 distance measure and linkage rule) until all points are combined into a single cluster—or until you decide to stop earlier.

In summary:

-Start with every point isolated in its own cluster.

-At each iteration, find the two clusters that are most similar.

-Merge them into a larger cluster.

-Keep repeating until one cluster remains (or your stopping condition is met).

This process builds a hierarchy of clusters, which you can slice at any level to get your desired number of groups.
"""
```

20.What makes Silhouette Score a better metric than just inertia for model evaluation

```python
"""
Silhouette Score goes further than inertia by weighing both cluster tightness and cluster separation.
While inertia only measures how close points are to their own centroid, Silhouette Score also considers
how far they are from other clusters. Because it takes both aspects into account, it offers a more well-rounded
evaluation of your clustering results.
"""
```

29.Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Generate synthetic data
blob_data, _ = make_blobs(n_samples=500, centers=3, random_state=7)

# Evaluate silhouette scores for different cluster counts
silhouette_scores = []
cluster_options = [2, 3, 4, 5]

for k in cluster_options:
    kmeans_model = KMeans(n_clusters=k, random_state=7)
    predicted_labels = kmeans_model.fit_predict(blob_data)
    score = silhouette_score(blob_data, predicted_labels)
    silhouette_scores.append(score)

# Display results with a bar plot
plt.bar(cluster_options, silhouette_scores, color='skyblue')
plt.title("Silhouette Scores for Different Cluster Counts")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.show()
```

30.Load the Iris dataset and use hierarchical clustering to group data. Plot a dendrogram with average linkage

31.Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with
decision boundaries

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

# Generate overlapping blobs
blob_data, _ = make_blobs(
    n_samples=400,
    centers=3,
    cluster_std=1.5,
    random_state=5
)

# Apply K‑Means clustering
kmeans_model = KMeans(n_clusters=3, random_state=5)
cluster_labels = kmeans_model.fit_predict(blob_data)

# Define the mesh grid for decision boundaries
step_size = 0.02
x_min, x_max = blob_data[:, 0].min() - 1, blob_data[:, 0].max() + 1
y_min, y_max = blob_data[:, 1].min() - 1, blob_data[:, 1].max() + 1
xx, yy = np.meshgrid(
    np.arange(x_min, x_max, step_size),
    np.arange(y_min, y_max, step_size)
)

# Predict labels for each point in the grid
grid_points = np.c_[xx.ravel(), yy.ravel()]
grid_labels = kmeans_model.predict(grid_points)
grid_labels = grid_labels.reshape(xx.shape)

# Plot decision boundaries and data points
plt.contourf(xx, yy, grid_labels, alpha=0.2, cmap='Accent')
plt.scatter(
    blob_data[:, 0],
    blob_data[:, 1],
    c=cluster_labels,
    cmap='Accent',
    edgecolor='k'
)
plt.title("K‑Means with Decision Boundaries")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

34.Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a
line plot

```python
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load and normalize the breast cancer dataset
cancer_dataset = load_breast_cancer()
scaled_features = StandardScaler().fit_transform(cancer_dataset.data)

# Compute inertia for different cluster counts
inertia_values = []
cluster_range = range(2, 7)

for num_clusters in cluster_range:
    kmeans_model = KMeans(n_clusters=num_clusters, random_state=0)
    kmeans_model.fit(scaled_features)
    inertia_values.append(kmeans_model.inertia_)

# Plot the elbow curve
plt.plot(cluster_range, inertia_values, marker='o')
plt.title("Inertia vs. Number of Clusters")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.grid(True)
plt.show()
```

42.Load the Breast Cancer dataset, reduce dimensionality using PCA, and apply Agglomerative Clustering.
Visualize in 2D

```python
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# Load and standardize the breast cancer dataset
cancer_dataset = load_breast_cancer()
scaled_features = StandardScaler().fit_transform(cancer_dataset.data)

# Reduce dimensionality via PCA
pca_transformer = PCA(n_components=2)
pca_embeddings = pca_transformer.fit_transform(scaled_features)

# Perform agglomerative clustering
agglomerative_model = AgglomerativeClustering(n_clusters=2)
cluster_labels = agglomerative_model.fit_predict(pca_embeddings)

# Visualize the clustering results
plt.scatter(
    pca_embeddings[:, 0],
    pca_embeddings[:, 1],
    c=cluster_labels,
    cmap='spring'
)
plt.title("Agglomerative Clustering on PCA-Reduced Cancer Data")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

43.Generate noisy circular data using make_circles and visualize clustering results from KMeans and DBSCAN
side-by-side

```python
from sklearn.datasets import make_circles
from sklearn.cluster import KMeans, DBSCAN
import matplotlib.pyplot as plt

# Create synthetic circular dataset
circle_data, _ = make_circles(n_samples=500, factor=0.4, noise=0.06)

# Perform K‑Means clustering
kmeans_model = KMeans(n_clusters=2, random_state=42)
kmeans_labels = kmeans_model.fit_predict(circle_data)

# Perform DBSCAN clustering
dbscan_model = DBSCAN(eps=0.2, min_samples=5)
dbscan_labels = dbscan_model.fit_predict(circle_data)

# Visualize both clustering results side by side
fig, axes = plt.subplots(ncols=2, figsize=(12, 5))
axes[0].scatter(circle_data[:, 0], circle_data[:, 1], c=kmeans_labels, cmap='coolwarm')
axes[0].set_title("K‑Means Clustering")

axes[1].scatter(circle_data[:, 0], circle_data[:, 1], c=dbscan_labels, cmap='coolwarm')
axes[1].set_title("DBSCAN Clustering")

plt.suptitle("K‑Means vs DBSCAN on Circular Data")
plt.show()
```

44.Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering

46.Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4
features)

```python
from sklearn.datasets import load_wine
from sklearn.cluster import KMeans
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt  # added since plot uses plt

# Load wine dataset into a DataFrame
wine_dataset = load_wine()
wine_df = pd.DataFrame(
    wine_dataset.data,
    columns=wine_dataset.feature_names
)

# Apply K‑Means clustering
kmeans_model = KMeans(n_clusters=3, random_state=42)
wine_df['cluster'] = kmeans_model.fit_predict(wine_df)

# Plot pairwise feature relationships colored by cluster
selected_features = wine_df.iloc[:, :4].join(wine_df['cluster'])
sns.pairplot(
    selected_features,
    hue='cluster',
    palette='husl'
)
plt.suptitle("Wine Dataset – K‑Means Pairplot", y=1.02)
plt.show()
```

47.Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the
count

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import numpy as np

# Generate synthetic data with noise
blob_data, _ = make_blobs(
    n_samples=400,
    centers=3,
    cluster_std=1.5,
    random_state=18
)

# Apply DBSCAN clustering
dbscan_model = DBSCAN(eps=0.9, min_samples=6)
cluster_labels = dbscan_model.fit_predict(blob_data)

# Compute cluster and noise counts
unique_labels = set(cluster_labels)
num_clusters = len(unique_labels) - (1 if -1 in unique_labels else 0)
noise_count = list(cluster_labels).count(-1)

print(f"Clusters found: {num_clusters}")
print(f"Noise points: {noise_count}")
```

48.Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the
clusters

```python
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# Load and reduce dimensionality of the digits dataset
digits_dataset = load_digits()
tsne_transformer = TSNE(n_components=2, random_state=42)
tsne_embeddings = tsne_transformer.fit_transform(digits_dataset.data)

# Perform agglomerative clustering on the reduced data
agglomerative_model = AgglomerativeClustering(n_clusters=10)
cluster_labels = agglomerative_model.fit_predict(tsne_embeddings)

# Visualize clusters in t-SNE space
plt.scatter(
    tsne_embeddings[:, 0],
    tsne_embeddings[:, 1],
    c=cluster_labels,
    cmap='nipy_spectral',
    s=30
)
plt.title("Agglomerative Clustering on Digits (t-SNE)")
plt.xlabel("t-SNE component 1")
plt.ylabel("t-SNE component 2")
plt.show()
```
